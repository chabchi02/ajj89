#### Missing values

**We first treat 'None' as a NaN value**

```python
toronto.replace('None', pd.NaT, inplace=True)
```

```python
toronto.isnull().sum()
```

**Deal missing values for 'ACCNUM'**

```python
# Grouping by year, date, time, and street
grouped = toronto.groupby(["YEAR", "DATE", "TIME", "STREET1"])

# Assigning the same ACCNUM to records with the same year, date, time, and street
for _, group in grouped:
    accnum = group["_id"].iloc[0]  # Taking the _id of the first record as ACCNUM
    nan_indices = group[group["ACCNUM"].isna()].index
    toronto.loc[nan_indices, "ACCNUM"] = accnum

toronto['ACCNUM'] = toronto['ACCNUM'].astype(int)
```

**Deal Null "FATAL_NO" as 0**

```python
toronto["FATAL_NO"] = toronto["FATAL_NO"].fillna(0).astype(int)
```

**Null value as "Unknown"**

```python
toronto["VISIBILITY"] = toronto["VISIBILITY"].fillna('Unknown')
```

```python
toronto["RDSFCOND"] = toronto["RDSFCOND"].fillna('Unknown')
```

```python
toronto["TRAFFCTL"] = toronto['TRAFFCTL'].fillna('Unknown')
```

```python
toronto["LOCCOORD"] = toronto["LOCCOORD"].fillna('Unknown')
```

**Null values as "No" for features that have values Yes/None**

```python
toronto['AUTOMOBILE'] = toronto['AUTOMOBILE'].fillna('No')
```

```python
toronto['PEDESTRIAN'] = toronto['PEDESTRIAN'].fillna('No')
```

**Null value as blank**

```python
toronto['STREET2'] = toronto['STREET2'].fillna('')
```

**Replace the rest of the Null value as "None"**

```python
toronto.replace(pd.NaT, 'None', inplace=True)
```

####Formatting

**Extracting Latitute and Longitude**

```python
toronto['geometry'] = toronto['geometry'].apply(lambda x: x.split("[[")[1].split("]]")[0].strip())
toronto[['Long', 'Lat']] = toronto['geometry'].str.split(',', n=1, expand=True)
toronto.drop(columns=['geometry'], inplace=True)
toronto
```

**Next is to format 'TIME' into String**

```python
# Function to convert time format
def convert_to_standard_time(time):
    time_str = str(time)
    return f"{time_str[:-2]}:{time_str[-2:]}"

# Apply the conversion function to the TIME column
toronto['TIME'] = toronto['TIME'].apply(convert_to_standard_time)
toronto.loc[toronto['TIME'].str.len() == 3, 'TIME'] =  "0" + toronto['TIME']
toronto.loc[toronto['TIME'].str.len() < 3, 'TIME'] =  "0:00"
toronto.head()
```

```python
toronto = toronto.reset_index(drop=True)
```

```python
toronto
```

**Have one accident incident on one row and not over multiple rows per injury**

```python
toronto = toronto.drop(toronto[toronto.INJURY == 'None'].index)
toronto = toronto.reset_index(drop=True)
#Have one accident incident on one row and not over multiple rows per injury
def accCategory_Fatal(row):
  if row['INJURY'] == 'Fatal':
    return 1
  else:
   return 0
def accCategory_Major(row):
  if row['INJURY'] == 'Major':
    return 1
  else:
   return 0
def accCategory_Minimal(row):
  if row['INJURY'] == 'Minimal':
    return 1
  else:
   return 0
def accCategory_Minor(row):
  if row['INJURY'] == 'Minor':
    return 1
  else:
   return 0

def accVehicle(row):
  if row['AUTOMOBILE'] == "Yes":
    return 1
  else:
    return 0

def accPedestrians(row):
  if row['PEDESTRIAN'] == "Yes":
    return 1
  else:
    return 0


toronto["Num_of_Vehicle"] = toronto.apply(accVehicle, axis = 1)
toronto['Num_Of_Pedestrians'] = toronto.apply(accPedestrians, axis = 1)
toronto['Num_of_Fatal_Injuries'] = toronto.apply(accCategory_Fatal, axis = 1)
toronto['Num_of_Major_Injuries'] = toronto.apply(accCategory_Major, axis=1)
toronto['Num_of_Minimal_Injuries'] = toronto.apply(accCategory_Minimal, axis=1)
toronto['Num_of_Minor_Injuries'] = toronto.apply(accCategory_Minor, axis=1)
aggregations = {
    'YEAR': 'first',
    'DATE': 'first',
    'TIME': 'first',
    'STREET1': 'first',
    'STREET2': 'first',
    'LOCCOORD': 'first',
    'TRAFFCTL': 'first',
    'VISIBILITY': 'first',
    'LIGHT': 'first',
    'RDSFCOND': 'first',
    'INJURY': 'first',
    'FATAL_NO': 'first',
    'PEDESTRIAN': 'first',
    'AUTOMOBILE': 'first',
    'NEIGHBOURHOOD_158': 'first',
    'Long': 'first',
    'Lat': 'first',
    "Num_of_Vehicle" : "sum",
    "Num_Of_Pedestrians" : "sum",
    'Num_of_Fatal_Injuries': 'sum',
    'Num_of_Major_Injuries': 'sum',
    'Num_of_Minimal_Injuries': 'sum',
    'Num_of_Minor_Injuries': 'sum'
}

toronto = toronto.groupby('ACCNUM').agg(aggregations).reset_index()
toronto = toronto.drop(columns=['INJURY', 'FATAL_NO', 'PEDESTRIAN', 'AUTOMOBILE'])

toronto
```

**Have one address for Toronto**

```python
toronto['STREET1'] = toronto['STREET1'] = toronto.apply(lambda row: row['STREET1'] + "@" + row['STREET2'] if not pd.isnull(row['STREET2']) else row['STREET1'], axis=1)
```

```python
def replace_month(value):
  months = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
  return months[value]

def replace_weekday(value):
  weekdays = {0:'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
  return weekdays[value]
# [T.T.] Cell can be only run once otherwise values are already replaced and throws error
date_ottawa['Month'] = date_ottawa['Month'].apply(replace_month)
date_ottawa['Day'] = date_ottawa['Day'].apply(replace_weekday)
date_ottawa
```

```python
def get_AMPM(value):
  if value >= 0 and value <= 11:
    return 'AM'
  if value >= 12 and value <= 23:
    return 'PM'

def to_notation_12(value):
  hour_minute = value.split(':', 1)
  minute = hour_minute[1]
  hour = int(hour_minute[0])
  if hour/12>1:
    conversion = hour % 12
    return("%d:%s" % (conversion, minute))
  elif hour == 0:
    hour_minute[0] = 12
    return("12:%s" % minute)
  else:
    return value

def get_quarter(value):
  if value == 'January' or value == 'February' or value == 'March':
    return '1st'
  elif value == 'April' or value == 'May' or value == 'June':
    return '2nd'
  elif value == 'July' or value == 'August' or value == 'September':
    return '3rd'
  elif value == 'October' or value == 'November' or value == 'December':
    return '4th'

# Should I include Holidays?
def is_business_day(date):
  if date == 'Saturday' or date == 'Sunday':
    return 'No'
  else:
    return 'Yes'

date_ottawa['AMPM'] = date_ottawa['Hour'].apply(get_AMPM)
date_ottawa['Notation12'] = date_ottawa['Time'].apply(to_notation_12)
date_ottawa['Quarter'] = date_ottawa['Month'].apply(get_quarter)
date_ottawa['is_Business_Day'] = date_ottawa['Day'].apply(is_business_day)
date_ottawa = date_ottawa[['Day', 'Time', 'Hour', 'Minute', 'AMPM', 'Notation12', 'Calendar_Date', 'Month', 'Year', 'Quarter', 'is_Business_Day']]
date_ottawa
```

## Emissions Fact

```python
def get_Date_Key(date, time):
    return collision_date.loc[(collision_date["Calendar_Date"] == date) & (collision_date["Time"] == time), 'Date_key'].iloc[0]
def get_Col_Env_Key(visibility, light, road_surface, traffic_control):
    return collision_envt.loc[(collision_envt["Visibility"] == visibility) & (collision_envt["Light"] == light) & (collision_envt["Road_surface"] == road_surface) & (collision_envt["Traffic_Control"] == traffic_control), 'Col_Env_key'].iloc[0]
ottawa["Location_key"] = range(0,len(ottawa))
ottawa['Date_key'] = ottawa.apply(lambda x: get_Date_Key(str(x['Accident_Date']), str(x['Accident_Time'])), axis=1)
ottawa['Col_Env_key'] = ottawa.apply(lambda x: get_Col_Env_Key(x['Environment_Condition'].lower(), x['Light'].lower(), x['Road_Surface_Condition'].lower(), x['Traffic_Control'].lower()), axis=1)
ottawa['Weather_key'] = range(0,len(ottawa))
ottawa.drop(columns=["Accident_Year", 'Accident_Date', 'Accident_Time', 'Location', 'Location_Type', 'Road_Surface_Condition', 'Environment_Condition', 'Light', 'Traffic_Control', 'Max_Injury', 'Lat', 'Long', 'Place_name', 'neighborhood_name', 'Num_of_Injuries'], inplace=True)
ottawa
```

```python
toronto["Location_key"] = range(787,len(toronto)+787)
toronto['Date_key'] = toronto.apply(lambda x: get_Date_Key(str(x['DATE']), str(x['TIME'])), axis=1)
toronto['Col_Env_key'] = toronto.apply(lambda x: get_Col_Env_Key(x['VISIBILITY'].lower(), x['LIGHT'].lower(), x['RDSFCOND'].lower(), x['TRAFFCTL'].lower()), axis=1)
toronto['Weather_key'] = range(787,len(toronto)+787)
toronto.drop(columns=["YEAR", 'DATE', 'TIME', 'STREET1', 'STREET2', 'LOCCOORD', 'TRAFFCTL', 'VISIBILITY', 'LIGHT', 'RDSFCOND', 'NEIGHBOURHOOD_158', 'Long', 'Lat', 'Place_name'], inplace=True)
toronto = toronto.rename(columns={'ACCNUM': 'ID', 'oldName2': 'newName2'})
toronto
```
